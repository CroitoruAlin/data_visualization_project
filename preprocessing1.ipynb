```python
#downloading the images of the players from the top 30 teams from www.sofifa.com using BeautifulSoup 
from bs4 import BeautifulSoup
from google.colab.files import download
from  requests import get
```

```python
teamsId=[45,73,240,243,9,48,18,10,5,21,44,11,241,1,2,22,461,46,47,100761,100757,112847,113180,110816,308,110815]
```

```python
for id in teamsId:
  u='https://sofifa.com/team/'+str(id)+"?r=200061&set=true"
  rsp=get(u)
  t=rsp.text
  bsoup=BeautifulSoup(t, 'html.parser')
  for div1 in bsoup.findAll('div', {"class":"lineup"}):
    for div2 in div1.findAll('div', {'class':'field-player'}):
        for (im,a) in zip(div2.findAll('img'),div2.findAll('a')):
            name='imagess/'+a["href"][15:-1]+".png"
            rsp=get(im['data-src']).content
            print(name)
            f=open(name, 'wb')
            f.write(rsp)
```
